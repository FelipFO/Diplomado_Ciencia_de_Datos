# **Introducción a la Estadística Descriptiva**


# Definición y objetivos

La estadística descriptiva es una rama de la estadística que se ocupa de recolectar, organizar, resumir y presentar un conjunto de datos de manera informativa. Su principal objetivo es proporcionar una descripción clara y comprensible de las características principales de un conjunto de datos, sin intentar inferir conclusiones que vayan más allá de los datos analizados.

Esto incluye la utilización de medidas de tendencia central, como la media, mediana y moda, así como medidas de dispersión, como el rango, la varianza y la desviación estándar, para resumir y describir las características esenciales de un conjunto de datos.

La estadística descriptiva también se apoya en herramientas gráficas como histogramas, diagramas de caja y gráficos de dispersión para visualizar y entender mejor la distribución y la relación entre los datos.

Su aplicación es fundamental en una variedad de campos, desde la investigación científica hasta la toma de decisiones en negocios, economía, ingeniería, entre otros.

# Diferencias entre estadística descriptiva e inferencial

Mientras que la estadística descriptiva se centra en describir y resumir los datos disponibles, la estadística inferencial toma un paso adicional al utilizar los datos de una muestra para hacer inferencias o generalizaciones sobre una población más grande. La estadística inferencial aplica métodos que permiten estimar parámetros, probar hipótesis y predecir tendencias basándose en la muestra de datos, asumiendo un cierto nivel de incertidumbre.

La principal diferencia radica, por tanto, en el propósito y en el alcance de los análisis. La estadística descriptiva no intenta alcanzar conclusiones que se extiendan más allá de los datos que se están analizando, es decir, se limita a lo que los datos dicen de sí mismos. Por otro lado, la estadística inferencial busca extrapolar los hallazgos de la muestra a la población mayor, proporcionando estimaciones y predicciones con un grado conocido de confianza.

En resumen, mientras la estadística descriptiva nos ofrece una "fotografía" de los datos disponibles, la estadística inferencial intenta "proyectar" esa fotografía más allá de los datos inmediatos, hacia conclusiones más amplias sobre lo que esos datos representan. Ambas son complementarias y esenciales en el análisis estadístico, proporcionando las herramientas necesarias para comprender los datos y tomar decisiones basadas en ellos.

# **Tipos de Datos y Escalas de Medición**

La comprensión de los diferentes tipos de datos y escalas de medición es fundamental en estadística, ya que determina el tipo de análisis estadístico que se puede realizar.

# Datos cualitativos vs cuantitativos

*   Datos cualitativos (o categóricos): Representan características o atributos y se pueden dividir en categorías distintas basadas en alguna cualidad inherente. Por ejemplo, los colores de los coches en un estacionamiento (rojo, azul, verde) son datos cualitativos.
*   Datos cuantitativos (o numéricos): Son medibles y se expresan numéricamente. Permiten operaciones matemáticas. Los datos cuantitativos se subdividen en discretos (valores enteros o contables, como el número de hijos en una familia) y continuos (valores que pueden tomar cualquier valor dentro de un rango, como el peso o la altura).



```
# Ejemplo de datos cualitativos y cuantitativos
datos_cualitativos = ['rojo', 'azul', 'verde', 'azul', 'rojo']
datos_cuantitativos_discretos = [1, 3, 2, 4, 3]
datos_cuantitativos_continuos = [1.73, 1.82, 1.65, 1.75, 1.80]

print("Datos cualitativos (categóricos):", datos_cualitativos)
print("Datos cuantitativos discretos:", datos_cuantitativos_discretos)
print("Datos cuantitativos continuos:", datos_cuantitativos_continuos)

```



# Escalas de medición: nominal, ordinal, intervalo, razón



*   Nominal: La escala nominal clasifica datos en categorías mutuamente excluyentes, sin ningún orden inherente. Por ejemplo, el tipo de vehículo (carro, camión, motocicleta) es una variable nominal.

*   Ordinal: Al igual que la nominal, clasifica datos en categorías, pero estas categorías tienen un orden o jerarquía. Un ejemplo sería la clasificación de un restaurante (1 estrella, 2 estrellas, 3 estrellas).

*   Intervalo: Esta escala tiene tanto orden como igualdad de intervalos entre las unidades, pero no tiene un verdadero cero. Un ejemplo sería la temperatura en grados Celsius.

*   Razón: Similar a la escala de intervalo, con la adición de un verdadero cero, lo que permite comparaciones de razones. Ejemplos incluyen el peso o la distancia.



```
# Ejemplo de escalas de medición
datos_nominales = ['carro', 'camión', 'motocicleta', 'carro']
datos_ordinarios = ['1 estrella', '2 estrellas', '3 estrellas', '2 estrellas']
datos_intervalo = [20, 22, 19, 21] # Temperatura en grados Celsius
datos_razon = [500, 1500, 700] # Peso en gramos

print("Datos nominales:", datos_nominales)
print("Datos ordinales:", datos_ordinarios)
print("Datos de intervalo:", datos_intervalo)
print("Datos de razón:", datos_razon)

```

La elección correcta de la escala de medición afecta a la validez de los métodos estadísticos aplicados. Por ejemplo, las medias y las desviaciones estándar no se deben calcular para datos nominales o ordinales, ya que estos datos no reflejan cantidades numéricas inherentes. En cambio, para esos tipos de datos, se utilizan frecuencias o medianas.

La comprensión y aplicación adecuada de estos conceptos es crucial para la interpretación correcta de los resultados del análisis estadístico.

Puntos clave a tener en cuenta:

Proporcionan una visión de la forma de la distribución de los datos.
Permiten identificar rápidamente los valores más comunes y las posibles anomalías.

Puntos clave a tener en cuenta:

Útiles para identificar relaciones o correlaciones entre variables.

# **Medidas de Tendencia Central**

Las medidas de tendencia central son estadísticas descriptivas que resumen con un único valor el centro de un conjunto de datos. Estas medidas son fundamentales porque nos dan una idea de los valores típicos o centrales en una distribución de datos. Las tres medidas más comunes de tendencia central son la media, la mediana y la moda.

# Media

Definición: La media es el promedio de todos los valores numéricos en un conjunto de datos.

Cálculo: Se calcula sumando todos los valores y dividiendo entre el número total de valores.

Importancia: La media es útil para determinar la ubicación central de los datos, pero puede ser sensible a valores atípicos extremos.

Puntos clave a tener en cuenta:

Proporciona un resumen numérico central de los datos.
Es sensible a valores atípicos, que pueden sesgar la media.



```
# .mean()
```



````python
#Codigo ilustrativo

import numpy as np

# Datos
datos = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
datos_1 = np.array([1, 2, 2, 3, 4, 7, 9]) #Conjunto de datos impar
datos_2 = np.array([1, 2, 3, 4, 6, 7]) #Conjunto de datos par


# Calcular la media
media = np.mean(datos_1)
print(f"Media: {media}")
````

# Mediana

Definición: La mediana es el valor medio que separa la mitad superior de la mitad inferior de un conjunto de datos cuando está ordenado.

Cálculo: Si el número de observaciones es impar, la mediana es el valor central. Si es par, es el promedio de los dos valores centrales.

Importancia: La mediana es menos sensible a valores atípicos y puede ser una mejor medida de la tendencia central para distribuciones muy asimétricas.

Puntos clave a tener en cuenta:

Divide el conjunto de datos en dos partes iguales.
Menos sensible a valores extremos que la media.

````python
#Codigo ilustrativo

# Calcular la mediana
mediana = np.median(datos_1)
mediana2 = np.median(datos_2)

#Imprimir valores
print(f"Mediana: {mediana}")
print(f"Mediana: {mediana2}")
````

# Moda

Definición: La moda es el valor o valores que aparecen con mayor frecuencia en un conjunto de datos.

Cálculo: Se identifica observando la frecuencia de cada valor.

Importancia: La moda es útil para datos categóricos y para comprender los valores comunes en un conjunto de datos.

Puntos clave a tener en cuenta:

Indica el valor más frecuente en un conjunto de datos.

Puede haber múltiples modas en un conjunto de datos.

Aplicable a datos cualitativos.

````python
# Seleccionar una columna para ejemplificar: 'salario_anual'
columna = 'salario_anual'

# Cálculo y visualización de la media, mediana y moda
media = df[columna].mean()
mediana = df[columna].median()
moda = df[columna].mode()[0]

print(f"Media: {media:.4f}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")

# Utilizamos .2f en la media para formatear la salida a dos decimales
````

# Medidas de Dispersión

Las medidas de dispersión son estadísticas descriptivas que indican qué tan esparcidos están los datos alrededor de una medida central. Proporcionan una comprensión de la variabilidad o diversidad dentro de un conjunto de datos. Las principales medidas de dispersión incluyen el rango, la varianza, la desviación estándar y el coeficiente de variación.

## Rango

Definición: El rango es la diferencia entre el valor máximo y mínimo en un conjunto de datos.

Cálculo: Se obtiene restando el valor mínimo del máximo.

Importancia: Proporciona una idea de la dispersión total de los datos, pero puede ser sensible a valores atípicos.

Puntos clave a tener en cuenta:

Simple de calcular y entender.
No toma en cuenta la distribución completa de los datos.
Código ilustrativo:



```
import numpy as np

# Datos
datos = np.array([1, 2, 5, 7, 9])

# Calcular el rango
rango = np.max(datos) - np.min(datos)
print(f"Rango: {rango}")

```


## Varianza

Definición: La varianza mide qué tan dispersos están los datos alrededor de la media.

Cálculo: Se calcula como el promedio de las diferencias al cuadrado entre cada valor y la media del conjunto de datos.

Importancia: Ofrece una medida precisa de la dispersión de los datos, aunque su unidad de medida no es la misma que la de los datos originales.

Puntos clave a tener en cuenta:

Proporciona una medida de la dispersión alrededor de la media.
Sus unidades son el cuadrado de las unidades de los datos, lo que puede complicar su interpretación.



```
# Calcular la varianza
varianza = np.var(datos)
print(f"Varianza: {varianza}")

```



## Desviación Estándar

Definición: La desviación estándar es la raíz cuadrada de la varianza.

Cálculo: Se obtiene tomando la raíz cuadrada de la varianza.

Importancia: Mide la dispersión de los datos en las mismas unidades que los datos originales, facilitando la interpretación.

Puntos clave a tener en cuenta:

Proporciona una medida de la dispersión alrededor de la media en las mismas unidades que los datos.
Más interpretable que la varianza para comparar la dispersión entre diferentes conjuntos de datos.



```
# Calcular la desviación estándar
desviacion_estandar = np.std(datos)
print(f"Desviación Estándar: {desviacion_estandar}")

```

Coeficiente de Variación
El coeficiente de variación (CV) es una medida relativa de la dispersión expresada como un porcentaje. Se calcula como la desviación estándar dividida por la media del conjunto de datos, multiplicada por 100. El CV es útil para comparar la variabilidad entre distribuciones con diferentes unidades de medida o medias.

Puntos clave a tener en cuenta:

Proporciona una medida relativa de la dispersión en términos porcentuales.
Útil para comparar la variabilidad entre distribuciones con diferentes medias o unidades.
Código ilustrativo:



```
# Calcular el coeficiente de variación
cv = (desviacion_estandar / np.mean(datos)) * 100
print(f"Coeficiente de Variación: {cv:.2f}%")

```

# Coeficiente de Variación

El coeficiente de variación (CV) es una medida de dispersión relativa que nos ayuda a entender la variabilidad de los datos en relación con la media, siendo especialmente útil para comparar la variabilidad entre distribuciones con medias distintas.

````python
def calcular_cv(data):
    media = data.mean()
    desviacion_std = data.std()
    cv = (desviacion_std / media) * 100
    return cv

cv_salario = calcular_cv(df['salario_anual'])
print(f"Coeficiente de Variación del Salario Anual: {cv_salario:.2f}%")
````

Este código define una función para calcular el coeficiente de variación, que luego se aplica a la columna salario_anual del DataFrame. El CV se expresa como un porcentaje, lo que facilita interpretar la variabilidad de los datos en relación con su media.

# Simetría (Skewness)

La simetría mide la asimetría de la distribución de los datos alrededor de su media.
Una distribución simétrica tiene una simetría de cero, mientras que una distribución sesgada a la derecha (colas hacia la derecha) tendrá una simetría negativa y una sesgada a la izquierda (colas hacia la izquierda), una simetría positiva.

````python
skewness = df['horas_trabajo_semanal'].skew()
print(f"Simetría del Salario Anual: {skewness:.2f}")
````

Este código calcula la simetría de la columna salario_anual. Un valor cercano a cero indica una distribución simétrica, mientras que valores positivos o negativos indican sesgo derecho o izquierdo, respectivamente.

# Curtosis

La curtosis mide la agudeza o achatamiento de la distribución de los datos en comparación con una distribución normal. Una curtosis positiva indica una distribución más puntiaguda (leptocúrtica), mientras que una negativa indica una más achatada (platicúrtica) que la normal.

````python
curtosis = df['salario_anual'].kurtosis()
print(f"Curtosis del Salario Anual: {curtosis:.2f}")
````

Este código calcula la curtosis para la columna salario_anual. Un valor cercano a cero sugiere una curtosis similar a la de una distribución normal, valores positivos indican una distribución más concentrada y valores negativos, una más dispersa.
