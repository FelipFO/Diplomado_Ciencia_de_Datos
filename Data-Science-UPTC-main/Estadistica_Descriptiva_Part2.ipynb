```python
columna_3 = 'Ingresos'

#Crear una funcion al objeto:

plt.figure(figsize=(10, 6))
sns.histplot(df[columna_3], bins=20, kde=True)
plt.title('Distribución de Ingresos')
plt.xlabel('Ingresos')
plt.ylabel('Frecuencia')
plt.show()
```

## Diagrama de Caja (Boxplot)

Uso: Representa gráficamente el resumen de cinco números (mínimo, primer cuartil, mediana, tercer cuartil y máximo) y los valores atípicos.
Importancia: Facilita la identificación de la simetría, dispersión y valores atípicos de los datos, permitiendo comparaciones rápidas entre diferentes grupos.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df[columna_3])
plt.title('Distribución de Ingresos')
plt.show()
```

```python
# Diagrama de dispersión entre 'Edad' e 'Ingresos'

plt.scatter(df['Edad'], df['Ingresos'], color='red')
plt.title('Relación entre Edad e Ingresos')
plt.xlabel('Edad')
plt.ylabel('Ingresos')
plt.show()
```
